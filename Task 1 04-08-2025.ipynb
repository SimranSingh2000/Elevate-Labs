```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
# 1. Check and handle missing values
print("Missing values before cleaning:\n", df.isnull().sum())
```

```python
# Example: fill missing 'country' and 'director' with 'Unknown'
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
```

```python
df.head()
```

```python
df.shape
```

```python
# 2. Remove duplicate rows
df = df.drop_duplicates()
```

```python
df.shape
```

```python
# 3. Standardize text fields
df['type'] = df['type'].str.strip().str.title()
df['country'] = df['country'].str.strip().str.title()
df['rating'] = df['rating'].fillna('Unrated').str.upper().str.strip()
```

```python
# 4. Convert date format to consistent datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
# 5. Rename columns to lowercase with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
# 6. Fix data types
df['release_year'] = df['release_year'].astype(int)
df['duration'] = df['duration'].astype(str).str.strip()
```

```python
# Preview cleaned data
print("\nCleaned Data Sample:\n", df.head())
```

```python
df.head()
```

```python
# Save the cleaned dataset
df.to_csv("netflix_titles_cleaned.csv", index=False)
```
